# NHL Stanley Cup Predictions Based on Historical Data

```python
import pandas as pd

regular_season = []
playoff_season = []

regular_file_list = ["NHL Regular Season History/Summary(" + str(i+1) + ").xlsx" for i in range(17)]
playoff_file_list = ["NHL Playoff History/Summary(" + str(i+1) + ").xlsx" for i in range(11)]

for filename in regular_file_list:
    df = pd.read_excel(filename, index_col=None, header=0)
    regular_season.append(df)
    
for filename in playoff_file_list:
    df = pd.read_excel(filename, index_col=None, header=0)
    playoff_season.append(df)

regular = pd.concat(regular_season, axis=0, ignore_index=True)
playoffs = pd.concat(playoff_season, axis=0, ignore_index=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE

# Remove duplicate entries from the regular DataFrame
regular = regular.drop_duplicates(subset=['Team', 'Season'])

# Create a new column 'Made Playoffs' in the regular DataFrame
regular['Made Playoffs'] = regular.apply(lambda row: 'Yes' if ((playoffs['Season'] == row['Season']) & (playoffs['Team'] == row['Team'])).any() else 'No', axis=1)

# Define the current season
current_season = 20232024

# Ensure all teams are included in the current season
all_teams = ['Washington Capitals', 'Chicago Blackhawks', 'Toronto Maple Leafs', 'Philadelphia Flyers', 'New York Rangers', 'New York Islanders', 'Pittsburgh Penguins', 'Boston Bruins', 'Buffalo Sabres', 'Montreal Canadiens', 'Ottawa Senators', 'Carolina Hurricanes', 'Florida Panthers', 'Tampa Bay Lightning', 'Detroit Red Wings', 'Columbus Blue Jackets', 'New Jersey Devils', 'St. Louis Blues', 'Nashville Predators', 'Colorado Avalanche', 'Minnesota Wild', 'Winnipeg Jets', 'Dallas Stars', 'Anaheim Ducks', 'Los Angeles Kings', 'San Jose Sharks', 'Calgary Flames', 'Edmonton Oilers', 'Vancouver Canucks', 'Arizona Coyotes', 'Vegas Golden Knights', 'Seattle Kraken']
current_season_teams = regular[regular['Season'] == current_season]['Team'].unique()
missing_teams = set(all_teams) - set(current_season_teams)

if missing_teams:
    # Add missing teams to the regular DataFrame with appropriate values
    missing_data = pd.DataFrame({'Team': list(missing_teams), 'Season': current_season, 'Made Playoffs': 'No', 'P%': 0, 'GF/GP': 0, 'GA/GP': 0, 'PP%': 0, 'PK%': 0})
    regular = pd.concat([regular, missing_data], ignore_index=True)

# Prepare the features (X) and target variable (y) from the regular season data
X_regular = regular[['P%', 'GF/GP', 'GA/GP', 'PP%', 'PK%']]
y_regular = regular['Made Playoffs']

# Convert '--' to NaN and then convert the data types
X_regular = X_regular.replace('--', np.nan)
X_regular = X_regular.apply(pd.to_numeric)

# Impute missing values using mean imputation
imputer = SimpleImputer(strategy='mean')
X_regular_imputed = imputer.fit_transform(X_regular)

# Apply SMOTE to balance the class distribution
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_regular_imputed, y_regular)

# Create a Random Forest Classifier for predicting playoff teams
playoff_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the playoff prediction model
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
playoff_model.fit(X_train, y_train)

# Predict playoff appearances for the current season based on the regular season data
current_season_data = regular[regular['Season'] == current_season]
current_season_features = current_season_data[['P%', 'GF/GP', 'GA/GP', 'PP%', 'PK%']]
current_season_features = current_season_features.replace('--', np.nan)
current_season_features = current_season_features.apply(pd.to_numeric)
current_season_features_imputed = imputer.transform(current_season_features)

current_season_probabilities = playoff_model.predict_proba(current_season_features_imputed)
predicted_playoff_teams = current_season_data.iloc[current_season_probabilities[:, 1].argsort()[::-1][:16]]['Team'].tolist()

print("Predicted playoff teams for the current season:")
print(predicted_playoff_teams)

# Create a new DataFrame with playoff teams and their regular season stats
playoff_teams_data = regular[(regular['Season'] == current_season) & (regular['Team'].isin(predicted_playoff_teams))]

# Prepare the features (X) and target variable (y) for predicting the Stanley Cup Winner
X_playoff = playoff_teams_data[['P%', 'GF/GP', 'GA/GP', 'PP%', 'PK%']]
y_playoff = playoff_teams_data['Team']

# Create a Random Forest Classifier for predicting the Stanley Cup Winner
winner_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the Stanley Cup Winner prediction model
winner_model.fit(X_playoff, y_playoff)

# Predict the probabilities for each playoff team to win the Stanley Cup
winner_probabilities = winner_model.predict_proba(X_playoff)

# Create a DataFrame with playoff teams and their probabilities of winning the Stanley Cup
probability_matrix = pd.DataFrame(winner_probabilities, columns=winner_model.classes_, index=predicted_playoff_teams)

print("\nProbability matrix for playoff teams to win the Stanley Cup:")
print(probability_matrix)

# Predict the Stanley Cup Winner
winner_prediction = winner_model.predict(X_playoff)

print("\nPredicted Stanley Cup Winner:")
print(winner_prediction[0])
```
